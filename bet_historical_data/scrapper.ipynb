```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time

# Configure Selenium WebDriver
options = Options()
options.binary_location = r"C:\Program Files\Opera\opera.exe"
options.add_argument("--headless")  # Run in headless mode
options.add_argument("--disable-gpu")
service = Service(r"C:\Users\Owner\Desktop\Betslip_Algo\bet_historical_data\operadriver_win64\operadriver.exe")  # Update with your chromedriver path

# Start the WebDriver
driver = webdriver.Chrome(service=service, options=options)

# Navigate to the webpage
url = "https://shop.sportybet.com/ng/liveResult/"
driver.get(url)
driver.implicitly_wait(10)

# Initialize the threshold score for Over/Under analysis
THRESHOLD_SCORE = 145  # Example threshold; adjust as needed

# Extract data
try:
    # Select the basketball category
    basketball_tab = driver.find_element(By.XPATH, "//button[text()='Basketball']")  # Adjust XPath as needed
    basketball_tab.click()
    time.sleep(3)  # Wait for the data to load

    # Extract game data
    games = driver.find_elements(By.CLASS_NAME, "gameId")  # Locate all games
    data = []
    for game in games:
        # Extract game ID and date
        game_id = game.get_attribute("textContent")
        game_date = game.find_element(By.CLASS_NAME, "time").text

        # Extract home and away teams
        home_team = game.find_element(By.CLASS_NAME, "home").text
        away_team = game.find_element(By.CLASS_NAME, "away").text

        # Extract scores
        scores = game.find_element(By.CLASS_NAME, "score-com").text.split("-")
        home_score = int(scores[0].strip())
        away_score = int(scores[1].strip())
        total_score = home_score + away_score

        # Determine Over/Under
        over_under = "Over" if total_score > THRESHOLD_SCORE else "Under"

        # Append extracted data
        data.append([game_id, game_date, home_team, away_team, home_score, away_score, total_score, over_under])

    # Create a DataFrame
    columns = ["Game ID", "Game Date", "Home Team", "Away Team", "Home Score", "Away Score", "Total Score", "Over/Under"]
    df = pd.DataFrame(data, columns=columns)

    # Save to CSV
    df.to_csv("basketball_historical_data.csv", index=False)
    print(df)

finally:
    # Close the WebDriver
    driver.quit()
```
